```python
## Create a loyalty flag for existing customers
# Split data into groups based on User_id and transform order_number to generate max orders for each user

df['max_order']= df.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
df.loc[df['max_order']>40, 'loyalty_flag']= 'Loyal Customer'
```

```python
df.loc[(df['max_order']<=40) & (df['max_order']>10), 'loyalty_flag']= 'Regular customer'
```

```python
df.loc[df['max_order']<=10, 'loyalty_flag']= 'New customer'
```

```python
#Does the price of products purchased by loyal customers differ from regular or new customers

df.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max', 'sum']})
```

#### New customers spend more on average than Loyal customers

```python
#split data into groups based on user_id, and transform prices to generate average_price for each user

df['average_price']= df.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
df.head(60)
```

```python
#derive column conditions

df.loc[df['average_price']<10, 'spending_flag']= 'Low spender'
```

```python
df.loc[df['average_price']>=10, 'spending_flag']= 'High spender'
```

```python
df['spending_flag'].value_counts(dropna= False)
```

```python
df[['customer_id', 'spending_flag', 'loyalty_flag', 'order_number']].head(20)
```

#split data into groups based on customer_ids, and transform days_since_prior_order to generate median_order_frequency

```python
df['median_order_frequency']= df.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
df.head(60)
```

```python
#derive column conditions

df.loc[df['median_order_frequency']>20, 'order_frequency_flag']= 'Non-frequent customer'
```

```python
df.loc[(df['median_order_frequency']>10) & (df['median_order_frequency']<=20), 'order_frequency_flag']= 'Regular customer'
```

```python
df.loc[df['median_order_frequency']<=10, 'order_frequency_flag']= 'Frequent customer'
```

```python
df['order_frequency_flag'].value_counts(dropna= False)
```
